```python
plt.figure(figsize = (8,6))


plt.plot(x, y, "-og")
plt.plot(x, y*0.4, "--r", linewidth=5)


plt.xlabel("x değişkeni", fontsize = 28)
plt.ylabel("y değişkeni", fontsize = 28)
plt.title("Sin Grafiği ", fontsize = 28)
plt.grid()

plt.legend(["Sin", "0.4*Sin"])

# plt.yticks([]) # y ekseninde değerleri siler
plt.xticks(ticks=[0, np.pi/2 ,np.pi, 2*np.pi], labels=[r"0",r"$\frac{\pi}{2}$", r"$\pi$", r"$2\pi$"], fontsize =20)


plt.annotate("Buluşma noktası \n \t ($\pi$,0)", [np.pi,0], [1.3*np.pi, 0.5], fontsize = 15, color = "blue",
            arrowprops = dict(shrink=0.02, facecolor="purple"))

plt.annotate("Buluşma noktası 2 \n  ($\pi$,0)", [np.pi/2,0.4], [0.5*np.pi, -0.5],
             fontsize = 15, color = "brown",
            arrowprops = dict(shrink=0.0002, facecolor="purple"),
             horizontalalignment = 'center',verticalalignment = 'top')
# Horizontalalignment  yazılan textin yatayda konumunu düzenler.  right, left center
# veritcalalignment yazılan textin dikeyde konumunu düzenler. top, center, bottom

plt.show();
```

```python
dataset = load_iris()

dataset.feature_names 
dataset.target
dataset['target_names'] 
```

```python
df =  pd.DataFrame(dataset.data, columns = dataset.feature_names ); df.head()
df['Species'] = dataset.target ; df.head()
```

```python
df.Species.value_counts()
```

```python
df.Species==0
```

```python
df['turler'] = np.where(df.Species==0, dataset.target_names[0], df.Species)
df['turler'] = np.where(df.Species==1, dataset.target_names[1], df.turler)
df['turler'] = np.where(df.Species==2, dataset.target_names[2], df.turler)
df.turler.values
```

```python
dataset.target_names[1]
```

```python
# Türlerin isimlerinden oluşan sütun oluşturalım.
df['Species1'] = df['Species'].apply(lambda x: dataset.target_names[x]) ; df.tail()
```

```python
# İstenilen şekilde kodlama yapalım
df['Species2'] = df.Species1.replace(to_replace = 'setosa', value= 'se')
df['Species2'] = df.Species1.replace(to_replace = 'versicolor', value= 've')
df['Species2'] = df.Species1.replace(to_replace = 'virginica', value= 'vi')
df.tail()
```

```python
df.drop(['Species', 'Species1','Species2'], axis=1)
```

```python
pd.factorize(df.Species1)
```

```python
codes, uniques = pd.factorize(df.Species1)
df['codes'] = codes  
df.tail()
```

```python
df1 = df.drop(['turler','Species','Species1','Species2'], axis=1)
df1.tail()
```

```python
df1.shape
df1.describe()
df1.info()
df1.isnull().sum()
```

```python
df1.duplicated().sum()
df1 = df1.drop_duplicates()
```

```python
df1.columns = ['sepal_length', 'sepal_width', 'petal_length','petal_width', 'codes']
df1.tail()
```
